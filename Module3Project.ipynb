```python
#grouped by zipcode
location = crime.groupby("ZIP")[["UCR CRIME CATEGORY"]]
```

```python
#grouped by time
datetime_of_crime = crime.groupby('OCCURRED ON')['UCR CRIME CATEGORY','ZIP']
```

```python
#Number of each crime type by time occurred
datetime_of_crime.value_counts()
```

```python
# Defined a function to classify crimes as violent or non-violent
def classify_crime(crime_type):
    violent_crimes = ['AGGRAVATED ASSAULT', 'ROBBERY', 'RAPE', 'MURDER AND NON-NEGLIGENT MANSLAUGHTER']
    return 'Violent' if crime_type in violent_crimes else 'Non-Violent'
```

```python
# Created a new column indicating whether each crime is violent or non-violent
crime['violent_category'] = crime['UCR CRIME CATEGORY'].apply(classify_crime)
```

```python
#Grouped by crime type
type_of_crime = crime.groupby(crime['violent_category'])['ZIP']
```

```python
#Non-violent crimes occurred more often
type_of_crime.count()
```
